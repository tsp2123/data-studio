```python
def vacation_maker(row):
    val = ''
    try:
        if re.findall(r'(2017-12-24).*', row['Pickup_Time']): 
            val = 'Tis_the_Season'
        elif re.findall(r'(2017-12-25).*', row['Pickup_Time']): 
            val = 'Tis_the_Season'
        elif re.findall(r'(2017-12-26).*', row['Pickup_Time']): 
            val = 'Tis_the_Season'
        elif re.findall(r'(2017-12-27).*', row['Pickup_Time']): 
            val = 'Tis_the_Season'
        elif re.findall(r'(2017-12-28).*', row['Pickup_Time']): 
            val = 'Tis_the_Season'
        elif re.findall(r'(2017-12-29).*', row['Pickup_Time']): 
            val = 'Tis_the_Season'
        elif re.findall(r'(2017-12-30).*', row['Pickup_Time']): 
            val = 'Tis_the_Season'
        elif re.findall(r'(2017-12-31).*', row['Pickup_Time']): 
            val = 'Tis_the_Season' 
        else:
            val = 'Workday'
    except:
        pass
    return val
    
```
