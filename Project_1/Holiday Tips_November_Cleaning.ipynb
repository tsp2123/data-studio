```python
def vacation_maker(row):
    if row[row['Pickup_Time'].str.contains(r'(2017-11-22).*')]: 
        val = 'Holiday'
    elif row[row['Pickup_Time'].str.contains(r'(2017-11-23).*')]: 
        val = 'Holiday'
    elif row[row['Pickup_Time'].str.contains(r'(2017-11-24).*')]: 
        val = 'Holiday'
    elif row[row['Pickup_Time'].str.contains(r'(2017-11-25).*')]: 
        val = 'Holiday'
    elif row[row['Pickup_Time'].str.contains(r'(2017-11-26).*')]: 
        val = 'Holiday'
    else:
        val = 'Workday'
    return val
        
```

```python
November['Type_of_Day'] = November.apply(vacation_maker,axis=1 )
```
