```python
colors = 'blue'
ax = US.groupby('Country').Number_Delivered.sum().sort_values(ascending=False).head(5).plot(kind='barh' , color=colors)


ax.set_title("US Military Exports", pad=40, fontsize=20)

ax.grid()


plt.savefig('US Exports Update.pdf')
plt.show()

```

```python
colors = 'orange'
ax = China.groupby('Country').Number_Delivered.sum().sort_values(ascending=False).head(5).plot(kind='barh' , color=colors)


ax.set_title("China Military Exports", pad=40, fontsize=20)
ax.grid()


plt.savefig('China Exports.pdf')
plt.show()

```
